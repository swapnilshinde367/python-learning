```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
```

# Uniform Distribution

```python
values = np.random.uniform(-10,10,10000)

plt.hist(values,50)

plt.show()
```

# Normal/Gaussian

Visualize the probability density function:

```python
x=np.arange(-3,3,0.001)

plt.plot(x,norm.pdf(x))
```
