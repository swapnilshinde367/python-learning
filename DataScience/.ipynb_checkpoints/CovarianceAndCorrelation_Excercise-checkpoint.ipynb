```python
%matplotlib inline

import numpy as np
from pylab import *


def def_mean( x ) :
    xmean = mean( x )
    return [xi-xmean for xi in x]

def def_covariance( x, y ) :
    n = len( x )
    return dot( def_mean( x ), def_mean( y )  ) / ( n  -1 )

pageSpeeds = np.random.normal( 3.0, 1.0, 1000 )
purchasesPerPage = np.random.normal( 50.0, 10.0, 1000 )

scatter( pageSpeeds, purchasesPerPage )

def_covariance( pageSpeeds, purchasesPerPage )
```

```python
# Fabricated purchases

purchasesPerPage = np.random.normal( 50.0, 10.0, 1000 ) / pageSpeeds
scatter( pageSpeeds, purchasesPerPage )

# Negative value shows less time taken to render the a page, more money spent
def_covariance( pageSpeeds, purchasesPerPage )
```

```python
# Correlation normalizes everything by giving their standard daviation, it gives values from -1 to 1 where ->
# -1 for perfect inverse correlation and 1 for perfect positive correlation

def def_correlation( x, y ) :
    stddevx = std( x )
    stddevy = std( y )
    return def_covariance( x, y ) / stddevx / stddevy

def_correlation( pageSpeeds, purchasesPerPage )
```

```python
# Numpy does it with function corrcoef which returns matrix of the correlation and coefficients
# between every combination of the arrays passed

np.corrcoef( pageSpeeds, purchasesPerPage )
```

```python
# Fabricating a linear purchase
purchasesPerPage = 100 - pageSpeeds * 3
scatter( pageSpeeds, purchasesPerPage )

print( def_covariance( pageSpeeds, purchasesPerPage ) )

# Numpy's covariance function
print( np.cov( pageSpeeds, purchasesPerPage ) )
```
