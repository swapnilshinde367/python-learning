```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

arrIncomes = np.random.normal( 25000, 15000, 10000 )
print ( 'mean before : ' + str( arrIncomes.mean() ))
print( 'histogram before' )
plt.hist( arrIncomes,50 )
plt.show()

arrIncomes = np.append( arrIncomes, [1000000000] )
print 'mean after : ' + str( arrIncomes.mean() )
print( 'histogram after' )
plt.hist( arrIncomes,50 )
plt.show()
```

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered

arrOutlieredIncomes = reject_outliers( arrIncomes )

print 'mean after : ' + str( np.mean( arrOutlieredIncomes ) )

plt.hist( arrOutlieredIncomes, 50 )
plt.show()
```
