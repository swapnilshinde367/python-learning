```python
%matplotlib inline
import numpy as np
from pylab import *
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

np.random.seed(2)
pageSpeeds = np.random.normal( 3.0, 1.0, 1000 )
purcahseAmounts = np.random.normal( 25.0, 2.5, 1000 ) / pageSpeeds

scatter( pageSpeeds, purcahseAmounts )
```

```python
x = np.array( pageSpeeds )
y = np.array( purcahseAmounts )

p4 = np.poly1d( np.polyfit( x, y, 5 ) )

# we wl visulaise scatter plot for the redicted values using polynoial for 0-7 seconds of pagespeeds
xp = np.linspace( 0, 7, 100 )

plt.scatter( x, y )
plt.plot( xp, p4(xp), c = 'y' )
plt.show()
```

```python
# r_squared_error

r2 = r2_score( y, p4( x ) )

r2
```
