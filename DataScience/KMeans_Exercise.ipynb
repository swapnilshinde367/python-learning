```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import array, random, float
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale

#Creating a fake data for income/age for intNumberOfPeople people in intNumberOfClusters clusters
def handleCreateFakeData( intNumberOfPeople, intNumberOfClusters ) :
    random.seed(10)
    fltPointsPerCluster = float (intNumberOfPeople)/intNumberOfClusters
    X = []
    for i in range( intNumberOfClusters ) :
        ageCentroid = random.uniform( 20.0, 70.0 )
        incomeCentroid = random.uniform( 2000.0, 20000.0 )
        
        for j in range(int(fltPointsPerCluster)) :
            X.append( [random.normal(incomeCentroid, 10000.0), random.normal( ageCentroid, 2.0 )] ) 
            
    return array( X )
        
data = handleCreateFakeData( 500, 4 )

model = KMeans( n_clusters = 8 )

#Scale data to normalize, for better results
model = model.fit( scale( data ) )

# print(model.labels_)

#visualize

plt.figure( figsize=( 8, 6 ) )
plt.scatter( data[:,0], data[:,1], c = model.labels_.astype(float))
plt.show()
```
