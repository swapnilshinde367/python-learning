```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from pylab import *
from sklearn.metrics import r2_score

pageSpeeds = np.random.normal( 3.0, 1.0, 100 )
purchaseAmounts = np.random.normal( 50.0, 30.0, 100 ) / pageSpeeds

scatter( pageSpeeds, purchaseAmounts)
```

```python
#Dividing 80% for training and 20% for testing
trainX = pageSpeeds[:80]
trainY = purchaseAmounts[:80]

testX = pageSpeeds[80:]
testY = purchaseAmounts[80:]


scatter( trainX, trainY )
```

```python
scatter( testX, testY )
```

```python
#let's plot against our training data
plot_train_x = np.array( trainX )
plot_train_y = np.array( trainY )

#8th degree polynomial
train_p4 = np.poly1d( np.polyfit( plot_train_x, plot_train_y, 8 ) )

xp = np.linspace( 0, 7, 100 )
axes = plt.axes()
axes.set_xlim( [0,7] )
axes.set_ylim( [0,200] )

plt.scatter( plot_train_x, plot_train_y )
plt.plot( xp, train_p4( xp ), c = 'y' )
plt.show()
```

```python
# Let's plot against test data
plot_test_x = np.array( testX )
plot_test_y = np.array( testY )

axes = plt.axes()
axes.set_xlim( [0,7] )
axes.set_ylim( [0,200] )
plt.scatter( plot_test_x, plot_test_y )

# We will use training p4.
plt.plot( xp, train_p4(xp), c='y' )
plt.show()
```

```python
train_r2 = r2_score( plot_train_y, train_p4(plot_train_x) )

train_r2
```

```python
test_r2 = r2_score( plot_test_y, train_p4(plot_test_x) )

test_r2
```
